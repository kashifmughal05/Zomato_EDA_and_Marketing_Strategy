### Rating count based on 'Aggregate rating', 'Rating color' and 'Rating text'

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

```python
final_df['Rating text'].value_counts()
```

```python
final_df['Rating text'].value_counts().sum()
```

## Observation
1. When ratings are between 4.5 to 4.9 ---> Excellent
2. When ratings are between 4.0 to 4.4 ---> Very Good
3. When ratings are between 3.5 to 3.9 ---> Good
4. When ratings are between 2.5 to 3.4 ---> Average
5. When ratings are between 1.8 to 2.4 ---> Poor

### Graphical Representation

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating count', data=ratings, palette='muted')
plt.title('No of ratings based on Aggregate rating values')
```

There is a high number with `No Rating`

```python
sns.barplot(x='Aggregate rating', y='Rating count', hue='Rating color', palette=['blue','red','orange','yellow','green','green'], data=ratings)
plt.title('No of ratings based on Aggregate rating values classified with Rating color')
```

## Observations:
1. Not Rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4

### Frequencies of each rating color in their ranges

```python
# Count plot
sns.countplot(x='Rating color', data=ratings, palette=['blue','red','orange','yellow','green','green'])
plt.title('No of ratings based on Rating color')
```
